```python
records = []
```

```python
def addDetails():
    athleteName = input("Enter athlete's name: ")
    country = input("Enter athlete's country: ")
    event = input("Enter event name: ")
    medal = input("Enter medal won (Gold, Silver, Bronze, or None): ")
    year = input("Enter the year: ")
    performance = input("Enter performance metrics: ")
    
    record = {
        'athleteName' : athleteName,
        'country': country,
        'event': event,
        'medal': medal,
        'year' : year,
        'performance': performance
    }
    
    records.append(record)
    print("Record added successfully...")
    
```

```python
def viewDetails():
    print(f"Performance in Last years:");
    for record in records:
        print(f"EventName   :{record['event']}")
        print(f"Year        :{record['year']}")
        print(f"Medal       :{record['medal']}")
        print(f"AtheleteName:{record['athleteName']}")
        print(f"Country     :{record['country']}")
        print(f"Performance :{record['performance']}")
        print(" ")
```

```python
def viewMedalCountByCountry():
    medalCount={};
    for record in records:
        year=record['year'];
        medal=record['medal'];
        country=record['country'];
        if year not in medalCount:
            medalCount[year]=0;
        if record['medal'] in ['Gold','Silver','Bronze']:
            medalCount[year]+=1;
    for YEAR,COUNT in medalCount.items():
        print(f"Country: {country} | Year: {YEAR} | Total Medal: {COUNT} \n" "");
```

```python
def searchAthletePerformance():
    athleteName=input("Enter Athlete Name: ")
    athleteRecords = [record for record in records if record['athleteName'].lower() == athleteName.lower()]

    if athleteRecords:
        for record in athleteRecords:
            print(f"Event: {record['event']} | Medal: {record['medal']} | Performance: {record['performance']}")
    else:
        print("No records found for this athlete.")
```

```python
def viewEventDetails():
    name=input("Enter the event name: ")

    for record in records:
        if record['event']==name:
            print(f"Event: {record['event']} |Year: {record['year']} |AthleteName: {record['athleteName']} |Meal: {record['medal']}\n") 
```

```python
def viewStatistics():
    totalRecords = len(records)
    if totalRecords == 0:
        print("No records available to generate statistics.")
        return
    
    avgPerformance = totalRecords / len(records)
    print(f"Total records: {totalRecords}")
    print(f"Average performance (number of entries): {avgPerformance}")
    viewMedalCountByCountry()
```

```python
def exitSystem():
    print("Exiting the system. Goodbye!")
    exit()
```

```python
def run():
    while True:
        print("\n--- Olympic Records Handling ---")
        print("1. Add Details")
        print("2. View Details")
        print("3. View Medal Count by Country")
        print("4. Search Athlete Performance")
        print("5. View Event Details")
        print("6. View Statistics")
        print("7. Exit")
        
        choice = input("Select an option: ")

        match choice:
            case '1':
                addDetails()
            case '2':
                viewDetails()
            case '3':
                viewMedalCountByCountry()
            case '4':
                searchAthletePerformance()
            case '5':
                viewEventDetails()
            case '6':
                viewStatistics()
            case '7':
                exitSystem()
            case _:
                print("Invalid option, please try again.")
            
            
run()
```
